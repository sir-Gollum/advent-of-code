```python
@dataclass
class Item:
    n: int
    marked: bool = False
    
        
class Board:
    def __init__(self, items: list[list[Item]]):
        self.items = items
        
    def __str__(self):
        return 'Board:\n%s' % ('\n'.join([
            ' '.join(['%2d' % (c.n,) for c in r])
            for r in self.items 
        ]))
    
    def items_str(self):
         return 'Board:\n%s' % ('\n'.join([
            ' '.join([str(c) for c in r])
            for r in self.items 
        ]))       

    def has_won(self):
        for r in self.items:
            if all((c.marked for c in r)):
                return True
                    
        for c_idx, _ in enumerate(self.items[0]):
            if all((r[c_idx].marked for r in self.items)):
                return True
        return False
    
    def mark(self, n: int):
        for r in self.items:
            for c in r:
                if c.n == n:
                    c.marked = True
        
    def sum_unmarked(self):
        return sum([
            c.n 
            for r in self.items
            for c in r
            if not c.marked
        ])
        
                
def parse(data):
    numbers = [int(n) for n in data[0].split(',')]    

    boards = []
    for board_data in data[1:]:
        board_items = [
            [Item(int(n)) for n in line.split()]
            for line in board_data.split('\n')            
        ]
#         print(board_items)
        board = Board(board_items)
        boards.append(board)
    return numbers, boards


def solve1(data):
    numbers, boards = parse(inp)
    
    for draw in numbers:
        print('Drawing', draw)
        for b in boards:
            b.mark(draw)
            if b.has_won():
                print('Winner:', b)                
                return b.sum_unmarked(), b.sum_unmarked() * draw
        
solve1(inp) 
    
# numbers, boards = parse(inp)
# print(numbers)
# print()
# print(boards[0])
# print(boards[0].has_won())
```
