The probe launcher on your submarine can fire the probe with any integer velocity in the x (forward) and y (upward, or downward if negative) directions. For example, an initial x,y velocity like 0,10 would fire the probe straight up, while an initial velocity like 10,-1 would fire the probe forward at a slight downward angle.

The probe's x,y position starts at 0,0. Then, it will follow some trajectory by moving in steps. On each step, these changes occur in the following order:

The probe's x position increases by its x velocity.
The probe's y position increases by its y velocity.
Due to drag, the probe's x velocity changes by 1 toward the value 0; that is, it decreases by 1 if it is greater than 0, increases by 1 if it is less than 0, or does not change if it is already 0.
Due to gravity, the probe's y velocity decreases by 1.
For the probe to successfully make it into the trench, the probe must be on some trajectory that causes it to be within a target area after any step. The submarine computer has already calculated this target area (your puzzle input). For example:

```
target area: x=20..30, y=-10..-5
```

If you're going to fire a highly scientific probe out of a super cool probe launcher, you might as well do it with style. How high can you make the probe go while still reaching the target area?

In the above example, using an initial velocity of 6,9 is the best you can do, causing the probe to reach a maximum y position of 45. (Any higher initial y velocity causes the probe to overshoot the target area entirely.)

Find the initial velocity that causes the probe to reach the highest y position and still eventually be within the target area after any step. What is the highest y position it reaches on this trajectory?

````python
from dataclasses import dataclass, field
import re
````

````python
@dataclass
class Target:
    minx: int
    maxx: int
    miny: int
    maxy: int
        
    def hit_status(self, x: int, y: int):
        if (
            self.minx <= x <= self.maxx 
            and self.miny <= y <= self.maxy
        ):
            # hit
            return 1
        
        if y < self.miny:
            # overshoot, stop
            return -1
        
        if self.minx < 0 and x < self.minx:
            # overshoot, stop
            return -1
        
        if self.maxx > 0 and x > self.maxx:
            # overshoot, stop
            return -1
        
        # do another step
        return 0
            
class Trajectory:
    vx: int
    vy: int 
    x: int = 0 
    y: int = 0
    dots: list[tuple[int, int]]
        
    def __init__(self, vx: int, vy: int):
        self.vx = vx
        self.vy = vy
        self.initial_vx = vx
        self.initial_vy = vy
        self.dots = [(0, 0)]
    
    def vis(self, t: Target):
        minx = min([d[0] for d in self.dots] + [t.minx])
        maxx = max([d[0] for d in self.dots] + [t.maxx])
        miny = min([d[1] for d in self.dots] + [t.miny])
        maxy = max([d[1] for d in self.dots] + [t.maxy])
#         print(f'Vis: x={minx}..{maxx}, y={miny}..{maxy}')
        dots_set = set(self.dots)
        
        res = []
        for y in range (maxy, miny-1, -1):
            row = []
            for x in range(minx, maxx+1):
                marker = '.'
                point = (x, y)
                is_hit = t.hit_status(*point) == 1               
                
                if point == (0, 0):
                    marker = 'S'
                elif point in dots_set:
                    marker = '#'
                elif is_hit:
                    marker = 'T'
                row.append(marker)
            res.append(''.join(row))
        return '\n'.join(res)
    
    def step(self):
        self.x += self.vx
        self.y += self.vy
            
        if self.vx:
            self.vx += -1 if self.vx > 0 else 1
        
        self.vy -= 1
        self.dots.append((self.x, self.y))
        
    def shoot(self, target: Target, log_hits: bool=False, log_misses: bool=False):
        for i in range(200):
            hit_st = target.hit_status(self.x, self.y)
            if hit_st == 1:
                if log_hits:
                    print(f'Step {i+1} hit!')
                    print(self.vis(target))
                    
                return True
            elif hit_st == -1:
                if log_misses:
                    print(f'Step {i+1} miss!')
                    print(self.vis(target))
                
                return False
            else:
                self.step()   
        return False
        

        
def solve12(data: str):
    minx, maxx, miny, maxy = list(map(int, re.sub(
        r'target area: x=([^.]+)\.\.([^.]+), y=([^.]+)\.\.([^.]+)', 
        r'\1 \2 \3 \4', 
        data
    ).split()))
    
    target = Target(minx, maxx, miny, maxy)
#     tr = Trajectory(6, 3)
#     tr.add_initial_dot()
    
    hitting_trajectories = []
    
    for vx in range(maxx+1):
        for vy in range(miny, 500):
            tr = Trajectory(vx, vy)
            if tr.shoot(target, log_hits=False):
                hitting_trajectories.append(tr)

    print('Solution 1:', max([max([d[1] for d in ht.dots]) for ht in hitting_trajectories]))
    distinct_trajectories = set([(ht.initial_vx, ht.initial_vy) for ht in hitting_trajectories])
    print('Solution 2:', len(distinct_trajectories))
#     print(distinct_trajectories)
    

# solve12("target area: x=20..30, y=-10..-5")
solve12("target area: x=281..311, y=-74..-54")
````
