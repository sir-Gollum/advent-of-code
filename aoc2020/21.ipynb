```python
from collections import defaultdict
import copy
import pprint
```

```python
ing_allergenes = defaultdict(set)

# collect ingredient -> possible allergenes
for row in rows:
    for ing in row[0]:
        for aller in row[1]:
            ing_allergenes[ing].add(aller)

cleaned_up = copy.deepcopy(dict(ing_allergenes))
print('Allergene candidates:')
pprint.pprint(cleaned_up)
print()

# drop possible allergene, if allergene is encountered in dishes without ingredient
for ing, possible_allergenes in ing_allergenes.items():
    for possible_aller in possible_allergenes:
        found = False
        for row in rows:
#             print('Check', ing, 'with', possible_aller, 'in', row[1], 'in', row[0])
            if possible_aller in row[1] and ing not in row[0]:
#                 print('Found!')
                found = True
                break
        if found:
            cleaned_up[ing].remove(possible_aller)

print('After removing impossible:')
pprint.pprint(cleaned_up)
print()

answer1 = 0
for ing, possible_aller in cleaned_up.items():
    if possible_aller:
        continue
        
    for row in rows:
        answer1 += row[0].count(ing)

print('Answer to #1:', answer1)
```

```python
deduced = copy.deepcopy(cleaned_up)

while any([len(v) > 1 for v in deduced.values()]):
    # round of deduction
    deduced_new = copy.deepcopy(deduced)
    
    allergenes_to_remove = set([
        list(a)[0] for a in 
        deduced.values()
        if len(a) == 1        
    ])
    
    print('\nRound: deducing allergenes:', sorted(allergenes_to_remove))
    pprint.pprint(deduced)
    
    for ing, allergenes in deduced.items():
        if len(allergenes) > 1:
            allergenes -= set(allergenes_to_remove)
            deduced_new[ing] = allergenes
            

print('\nDone:')
pprint.pprint(deduced)    
```
