```python
class Tile:
    def __init__(self, lines, trans=None):
        name = lines[0]
        content = lines[1:]
        self.tile_id = int(name.replace('Tile ', '').replace(':', ''))
        self.content = content
        self.trans = trans
        
        self.left = None
        self.top = None
        self.right = None
        self.bottom = None            
        
        self.top_row = self.content[0]
        self.bottom_row = self.content[-1]
        self.right_col = ''.join([l[-1] for l in self.content])
        self.left_col = ''.join([l[0] for l in self.content])
        
        self.matches = set()
        

    def transform(self, trans, content):
        return Tile(
            [
                f'Tile {self.tile_id}:'
            ] + content,
            trans
        )
        
    def __str__(self):
        return (
            f'<Tile {self.tile_id}:' + 
#             3'\n' + "\n".join(self.content) + 
            '>' + 
            (f'({self.trans})' if self.trans else '')
#             +
#             '\n'
        )
        
    def __repr__(self):
        return str(self)
    
    def vflip(self):
        return self.transform('vf', [
            ''.join(reversed(line))
            for line in self.content
        ])
    
    def hflip(self):
        return self.transform('hf', [
            line
            for line in reversed(self.content)
        ])
    
    def rotate(self, times=1):
        x = np.array([list(l) for l in self.content], dtype=object) 
        r = [''.join(i) for i in np.rot90(x, k=-times).tolist()]
        return self.transform('r%s' % (times,), r)
        
    
    def match(self, o):        
        if self.top_row == o.bottom_row:
#             assert not self.top and not o.bottom, (self, o)
#             self.top = o
#             o.bottom = self            
            return True, 't'
        
        if self.bottom_row == o.top_row:
#             assert not self.bottom and not o.top, (self, o)
#             self.bottom = o
#             o.top = self
            return True, 'b'
        
        if self.left_col == o.right_col:
#             assert not self.left and not o.right, (self, o)
#             self.left = o
#             o.right = self
            return True, 'l'
        
        if self.right_col == o.left_col:
#             assert not self.right and not o.left, (self, o)
#             self.right = o
#             o.right = self
            return True, 'r'
        
        return False, ''
            
    
def viz(topleft):
    t = topleft
    drow = [t]
    while t:
        while t.right:
            drow.append(t.right)
            t = t.right

        for content_items in zip(*[[f'Tile {t.tile_id}'] + t.content for t in drow]):
            print(' '.join(content_items))
        
        t = t.bottom
            

inp = [Tile(c) for c in inp_chunks]


for t in inp:
    for o in inp:
        if o == t:
            continue 
            
        if (
            t.match(o)[0]
            or t.match(o.vflip())[0]
            or t.match(o.hflip())[0]
            or any(t.match(o.rotate(i))[0] for i in range(1, 4))
            or any(t.match(o.vflip().rotate(i))[0] for i in range(1, 4))
            or any(t.match(o.hflip().rotate(i))[0] for i in range(1, 4))
        ):
            t.matches.add(o)

tiles2 = [
    t for t in inp if len(t.matches) == 2
]

print(tiles2)

t = tiles2[0].tile_id
for i in tiles2[1:]:
    t *= i.tile_id
print('Answer:', t)
```
