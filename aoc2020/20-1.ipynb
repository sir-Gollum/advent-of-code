```python
import numpy as np
import math
from collections import namedtuple
import copy


class Tile:
    def __init__(self, lines, trans=''):
        name = copy.deepcopy(lines[0])
        content = copy.deepcopy(lines[1:])
        self.tile_id = int(name.replace('Tile ', '').replace(':', ''))
        self.content = content
        self.trans = trans
        
        self.left = None
        self.top = None
        self.right = None
        self.bottom = None            
        
        self.top_row = self.content[0]
        self.bottom_row = self.content[-1]
        self.right_col = ''.join([l[-1] for l in copy.deepcopy(self.content)])
        self.left_col = ''.join([l[0] for l in copy.deepcopy(self.content)])
        
        self.matches = set()
        

    def transform(self, trans, content):
        return Tile(
            [
                f'Tile {self.tile_id}:'
            ] + copy.deepcopy(content),
            '+'.join([t for t in [self.trans, trans] if t])
        )
        
    def __str__(self):
        return (
            f'<Tile {self.tile_id}' + 
#             3'\n' + "\n".join(self.content) + 
            (f'({self.trans})' if self.trans else '') + 
            '>'

#             +
#             '\n'
        )
        
    def __repr__(self):
        return str(self)
    
    def __gt__(self, other):
        return self.tile_id > other.tile_id
    
    def vflip(self):
        return self.transform('vf', [
            ''.join(reversed(line))
            for line in self.content
        ])
    
    def hflip(self):
        return self.transform('hf', [
            line
            for line in reversed(self.content)
        ])
    
    def rotate(self, times=1):
        x = np.array([list(l) for l in copy.deepcopy(self.content)], dtype=object) 
        r = [''.join(i) for i in np.rot90(x, k=-times).tolist()]
        return self.transform('r%s' % (times,), r)
        
    def noborder(self):
        content = [
            line[1:-1]
            for line in self.content[1:-1]
        ]
        return self.transform('b', content)
    
    def match(self, o):        
        if self.tile_id == o.tile_id:
            return Match(False, '', o, None)
        
        if self.top_row == o.bottom_row:
            return Match(True, 't', o, o.tile_id)
        
        if self.bottom_row == o.top_row:
            return Match(True, 'b', o, o.tile_id)
        
        if self.left_col == o.right_col:
            return Match(True, 'l', o, o.tile_id)
        
        if self.right_col == o.left_col:
            return Match(True, 'r', o, o.tile_id)
        
        return Match(False, '', o, None)

    def all_matches(self, all_tiles, dedup=True):
        res = set()
        for o in all_tiles:
            if self == o:
                continue
                
            matches = [
                self.match(o),
                self.match(o.vflip()),
                self.match(o.hflip()),
            ] + [
                self.match(o.rotate(i)) for i in range(1, 4)
            ] + [
                self.match(o.vflip().rotate(i)) for i in range(1, 4)
            ] + [
                self.match(o.hflip().rotate(i)) for i in range(1, 4)
            ]

#             match = next((m for m in matches if m[0]), None)            
#             if match:
#                 res.add(match)        
            matches = [m for m in matches if m.is_match]
            if matches:
                res |= set(matches)                
                
        if dedup:
            result = {}
            for m in res:
                # dedup & choose simplest transformation
                if m.other_id not in result or len(m.other.trans) < len(result[m.other_id].other.trans):
                    result[m.other_id] = m


            return set(result.values())
        return res
    
    

Match = namedtuple('Match', ['is_match', 'type', 'other', 'other_id'])
SM = namedtuple('M', ['type', 'other'])

def to_sm(matches):
    return sorted([
        SM(m.type, m.other)
        for m in matches
    ])



    
def viz(topleft):
    t = topleft
    drow = [t]
    while t:
        while t.right:
            drow.append(t.right)
            t = t.right

        for content_items in zip(*[[f'Tile {t.tile_id}'] + t.content for t in drow]):
            print(' '.join(content_items))
        
        t = t.bottom
            

inp_chunks = [t.splitlines() for t in raw_inp.strip().split('\n\n')]

inp = [Tile(c) for c in inp_chunks]


for t in inp:
    t.matches = t.all_matches(inp)

corners = [
    t for t in inp if len(t.matches) == 2
]

print(corners)

answer = corners[0].tile_id
print(corners[0], [(m.type, m.other) for m in corners[0].matches])

for i in corners[1:]:
    print(i, [(m.type, m.other_id) for m in i.matches])
    answer *= i.tile_id
print('Answer:', answer)



# t1 = corners[0]
# t2 = [t for t in inp if t.tile_id == 2729][0]

# viz(t1), print(), viz(t2)
# t1.match(t2)
```

```python
def build_image(tl, all_tiles, corners):
    side = int(math.sqrt(len(all_tiles)))
    
    image = [[None for _ in range(side)] for _ in range(side)]
    image[0][0] = tl
    used = {tl.tile_id}
    
    for row in range(side):
        for col in range(side):
#             print('=' * 10 + f'\nStarting row {row}, col {col}, image: {image}')
            if row == 0 and col == 0:
#                 print('skip seed')
                continue            
    
            left_matches = set()
            left = None
            if col:                
                left = image[row][col-1]
                left_matches = [m for m in left.all_matches(all_tiles) if m.type == 'r' and m.other.tile_id not in used]
#             else:
#                 print('not looking for left matches')
            
            top_matches = set()
            top = None
            if row:
                top = image[row-1][col]
                top_matches = [m for m in top.all_matches(all_tiles) if m.type == 'b' and m.other.tile_id not in used]
#             else:
#                 print('not looking for top matches')
                
            
#             print(f'L matches (L is {left}):\n', left_matches, f'\nT matches (T is {top})\n', top_matches)
            if left_matches and top_matches:
                left_ids = set([m.other_id for m in left_matches])
                both_matches = set([m for m in top_matches if m.other.tile_id in left_ids])
                if len(both_matches) > 1:
#                     print(f'Too many matches for row {row} col {col}: {both_matches}')
                    return image
                if not both_matches:
#                     print(f'Cannot build image: no match at row {row} col {col}')
                    return image
                image[row][col] = list(both_matches)[0].other
            elif row == 0:
                lm = left_matches
                if not lm:
#                     print(f'Cannot build image: no left match at row {row} col {col}, all left matches: {left_matches}')
                    return image

                if len(lm) > 1:
#                     print(f'Too many LEFT matches for row {row} col {col}: {lm}')
                    return image
                
                image[row][col] = lm[0].other
                used.add(lm[0].other.tile_id)

            elif col == 0:
                tm = top_matches
                if not tm:
#                     print(f'Cannot build image: no top match at row {row} col {col}, all top matches: {top_matches}')
                    return image

                if len(tm) > 1:
#                     print(f'Cannot build image: too many BOTTOM matches for row {row} col {col}: {tm}')
                    return image
                    
                image[row][col] = list(tm)[0].other
                used.add(tm[0].other.tile_id)
    

    return image


def find_image(corners, all_tiles):
    for corner in corners:
        for transformed_corner in [
            corner,
            corner.vflip(),
            corner.hflip(),
        ] + [
            corner.rotate(i) for i in range(1, 4)
        ] + [
            corner.vflip().rotate(i) for i in range(1, 4)
        ] + [
            corner.hflip().rotate(i) for i in range(1, 4)
        ]:
            img = build_image(transformed_corner, all_tiles, corners)

            if all(sum(img, [])):
                return img

image_tiles = find_image(corners, inp)        
image_tiles
```

```python
class Image(Tile):
    def __init__(self, tiles=None, content=None):
        lines = ['Tile 666']
        
        if tiles:
            for row in tiles:
                for tile_line_no, _ in enumerate(row[0].noborder().content):
                    line = ''.join([t.noborder().content[tile_line_no] for t in row])
                    lines.append(line)
        else:
            lines += content
            
        super().__init__(lines=lines)    
    
    def transform(self, trans, content):       
        return Image(content=copy.deepcopy(content))

    
    
    def roughness(self):
        return sum([1 for line in self.content for c in line if c == '#'])
            
    def with_monsters(self):
        lines = copy.deepcopy(self.content)
#         print()
        for n in range(len(lines)-2):
            chunk = lines[n:n+3]
            h, b, f = chunk
            
            if re_head.search(h) and re_body.search(b) and re_foot.search(f):
#                 print('Monser found, n=%s:\n%s' % (n, '\n'.join(lines[n:n+3])))
#                 print()
                
                for m in re_body.finditer(b):
                    start, stop = m.span()
                    hh, bb, ff = list(h[start:stop]), list(b[start:stop]), list(f[start:stop])
                    
                    for c in range(len(hh)):
                        if pattern_head[c] == '#':
                            hh[c] = 'O'
                        if pattern_body[c] == '#':
                            bb[c] = 'O'
                        if pattern_foot[c] == '#':
                            ff[c] = 'O'
                    
                    h, b, f = list(h), list(b), list(f)
                    length = len(h)
                    
                    h = h[:start] + hh + h[stop:]
                    b = b[:start] + bb + b[stop:]
                    f = f[:start] + ff + f[stop:]
                    assert len(h) == length
                    h, b, f = ''.join(h), ''.join(b), ''.join(f)
                    
            lines[n:n+3] = [h, b, f]
        
        return Image(content=lines)
        
  
img = Image(image_tiles)

roughness_record = []

for n, trans_img in enumerate([
    img,
    img.vflip(),
    img.hflip(),
] + [
    img.rotate(i) for i in range(1, 4)
] + [
    img.vflip().rotate(i) for i in range(1, 4)
] + [
    img.hflip().rotate(i) for i in range(1, 4)
]):
    if trans_img.roughness() == trans_img.with_monsters().roughness():
        continue
    
    print('Found monsters in transformation', n)
    roughness_record.append(trans_img.with_monsters().roughness())


roughness_record
# img = Image(image_tiles)
# viz(img.hflip().rotate())
# # img.roughness()
# viz(img.hflip().rotate().with_monsters())
# img.hflip().rotate().with_monsters().roughness()
```
