```python
class Node:
    def __init__(self, color):
        self.cl = color
        self.contains = set()
        self.contained_in = set()

    def __lt__(self, other):
        return self.cl < other.cl
    
    def __repr__(self):
        return str(self)
    
    def __str__(self):
        return f'<Bag("{self.cl}"){[(x[0], x[1].cl) for x in sorted(self.contains)]}>'
        
x = Node('shiny gold')
x.contains.add((2, Node('test')))
n = Node('test3')
x.contains.add((3, n))
x.contains.add((4, n))
```

```python
ALL_NODES = {}

def get_node(color):
    if color in ALL_NODES:
        return ALL_NODES[color]
    else:
        node = Node(color)
        ALL_NODES[color] = node        
        return node
    

for line in inp:
    parts = line.split(' contain ')
    container_cl = parts[0].replace(' bags', '')
    container = get_node(container_cl)
    
    
    content = parts[1].split(', ')
    
    for it in content:
        if 'no other bags' in it:
            continue
        
        it = it.replace(' bags.', '').replace(' bag.', '').replace(' bags', '').replace(' bag', '').split()
        it_num = int(it[0])
        it_color = f'{it[1]} {it[2]}'
        
        it_node = get_node(it_color)
        container.contains.add((it_num, it_node))
        
        it_node.contained_in.add(container)
        
#     print(line)
#     print(container)
#     print()
```

```python
def find(start, containers=None):
    if containers is None:
        containers = set()
        
    for container in start.contained_in:
        containers |= {container}
        containers |= find(container, containers)

    return containers
    
    
len(find(ALL_NODES['shiny gold']))
```
