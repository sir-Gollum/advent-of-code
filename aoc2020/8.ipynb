```python
def terminates(program):
#     print('program:', program)
    seen = set()
    acc = 0
    ii = 0
    
    while ii not in seen:
#         print(inp[ii], '|', acc)        
        if ii == len(program):
            print('Acc:', acc)
            return True
        
        seen.add(ii)
        cmd, arg = program[ii].split()
        arg = int(arg)

        if cmd == 'acc':
            acc += arg
            ii += 1

        elif cmd == 'jmp':
            ii += arg

        elif cmd == 'nop':
            ii += 1
            
#     print(ii, seen)
    return False


def patch_next(program, end_idx):
    for i in range(end_idx, -1, -1):
        cmd, arg = program[i].split()
        if cmd == 'jmp':
            program[i] = f'nop {arg}'
            return program, i
        if cmd == 'nop':
            program[i] = f'jmp {arg}'
            return program, i
    
    return program, i        

            
modified_idx = len(inp) - 1 

while modified_idx >= 0:
    modified, modified_idx = patch_next(inp[:], modified_idx)
    print('Checking idx', modified_idx)
    if terminates(modified):
        print("Solution found!")
        break
    modified_idx -= 1
```
