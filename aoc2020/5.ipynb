```python
def decode(t):
    row_l, row_r = 0, 127
    for c in t[:-3]:
        half = (row_r - row_l) // 2
        if c == 'F':
            row_r = row_l + half
        else:
            row_l = row_l + half + 1
        
#         print(c, row_l, row_r)
    
    assert row_l == row_r
    
    col_l, col_r = 0, 7
    for c in t[-3:]:
        half = (col_r - col_l) // 2
        if c == 'L':
            col_r = col_l + half
        else:
            col_l = col_l + half + 1
        
#         print(c, col_l, col_r)
        
    assert col_l == col_r

    return row_l * 8 + col_l 
    
    
decode('BBFFBBFRLL')
```
