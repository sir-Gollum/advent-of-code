```python
from collections import defaultdict

spoken = []
mem = defaultdict(list)
turn = 0


for n, i in enumerate(inp):
    turn = n+1
    assert i not in mem
    
    mem[i].append(turn)
    spoken.append(i)


while turn < 2021:
    turn += 1    
    last_spoken = spoken[-1]
    new_spoken = mem[last_spoken][0] - mem[last_spoken][-1]
    
    if turn < 200:
        print('T', turn, 'speaking', new_spoken, 'mem', dict(mem))
    mem[new_spoken].insert(0, turn)
    mem[new_spoken] = mem[new_spoken][:2]
    spoken.append(new_spoken)
    
    
```
