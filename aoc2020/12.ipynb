```python
from collections import namedtuple

instr = namedtuple('instr', ('action', 'arg'))

def parse(s):
    res = []
    for line in s.strip().splitlines():
        res.append(instr(line[0], int(line[1:])))
    return res
```

```python
x = 0
y = 0

facing = 'E'


for i in inp:
    if i.action == 'N':
        y -= i.arg
    elif i.action == 'S':
        y += i.arg
    elif i.action == 'E':
        x += i.arg
    elif i.action == 'W':
        x -= i.arg
    elif i.action == 'R':
        if i.arg % 90 != 0:
            print(i)
            raise ValueError()
        
        times = i.arg // 90
        for _ in range(times):
            facing = {
                'N': 'E',
                'E': 'S',
                'S': 'W',
                'W': 'N',
            }[facing]    
    elif i.action == 'L':
        if i.arg % 90 != 0:
            print(i)
            raise ValueError()
        
        times = i.arg // 90
        for _ in range(times):
            facing = {
                'N': 'W',
                'W': 'S',
                'S': 'E',
                'E': 'N',
            }[facing]    
    elif i.action == 'F':
        if facing == 'E':
            x += i.arg
        elif facing == 'W':
            x -= i.arg
        elif facing == 'S':
            y += i.arg
        elif facing == 'N':
            y -= i.arg
    

print(x, y, 'answer:', abs(x)+abs(y))
```
