```python
import re


template = {
    'byr', # (Birth Year)
    'iyr', # (Issue Year)
    'eyr', # (Expiration Year)
    'hgt', # (Height)
    'hcl', # (Hair Color)
    'ecl', # (Eye Color)
    'pid', # (Passport ID)
    # 'cid', # (Country ID)
}

def is_valid(p):
    if len(set(p).intersection(template)) != len(template):
#         print('invalid - missing fields', template - set(p), p)
        return False

    t = p['byr']
    if not re.match(r'^\d{4}$', t):
        return False
    if not (1920 <= int(t) <= 2002):
#         print('invalid byr', t)
        return False

    t = p['iyr']
    if not re.match(r'^\d{4}$', t):
        return False    
    if not (2010 <= int(t) <= 2020):
#         print('invalid iyr', t)
        return False

    t = p['eyr']
    if not re.match(r'^\d{4}$', t):
        return False    
    if not (2020 <= int(t) <= 2030):
#         print('invalid eyr', t)
        return False

    t = p['hgt']
    m = re.match(r'^(\d+)(cm|in)$', t)
    if not m:
#         print('invalid hgt - no match', t)
        return False

    hgt, dim = m.groups()
    if dim == 'in':
        if not (59 <= int(hgt) <= 76):
#             print('invalid hgt in:', t)
            return False
    
    if dim == 'cm' :
        if not (150 <= int(hgt) <= 193):
#             print('invalid hgt cm:', t, '||', hgt, dim)
            return False

    t = p['hcl']
    m = re.match('^#[0-9a-f]{6}$', t)
    if not m:
#         print('invalid hcl - no match', t)
        return False

    if p['ecl'] not in {'amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth'}:
#         print('invalid tcl', p['ecl'])
        return False

    if not re.match(r'^\d{9}$', p['pid']):
#         print('invalid pid - no match', p['pid'])
        return False

    return True

valid = 0
cur = {}
for line in inp:
    if line.strip() == '':
        if cur and is_valid(cur):
            valid += 1

        cur = {}
    else:
        items = line.split()
        for it in items:
            k, v = it.split(':')
            cur[k] = v
else:
    if cur and is_valid(cur):
        valid += 1

# 180 - too high
valid
```
