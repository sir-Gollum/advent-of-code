```python
P1_WINS = 'p1_wins'
P2_WINS = 'p2_wins'

dt = datetime.datetime.now()


def play_game(p1_deck, p2_deck, game, stack):
    print(f'=== Game {game} ===')
    
    rnd = 0
    sub_game = 0
    while p1_deck and p2_deck:
        rnd += 1
        #print(f'\n-- Round {rnd} (Game {game}) --')
        #print(f"Player 1's deck: {p1_deck}")
        #print(f"Player 2's deck: {p2_deck}")
        
        stack_key = (tuple(p1_deck), tuple(p2_deck))
        if stack_key in stack:
            print('Recursion break!')
            if game > 1:
                return P1_WINS            
            return p1_deck
        
        stack |= {stack_key}

        p1 = p1_deck.pop(0)
        p2 = p2_deck.pop(0)
        #print(f'Player 1 plays: {p1}')
        #print(f'Player 2 plays: {p2}')
        
        if len(p1_deck) >= p1 and len(p2_deck) >= p2:
            #print('Playing a sub-game to determine the winner...\n')
            sub_game += 1
            winner = play_game(deepcopy(p1_deck)[:p1], deepcopy(p2_deck)[:p2], game + sub_game, set())
            #print(f'...anyway, back to game {game}.')
        else:
            winner = P1_WINS if p1 > p2 else P2_WINS
            
        if winner == P1_WINS:
            #print(f'Player 1 wins round {rnd} of game {game}!')
            p1_deck.extend([p1, p2])
        else:
            #print(f'Player 2 wins round {rnd} of game {game}!')
            p2_deck.extend([p2, p1])
            
        if (datetime.datetime.now() - dt).total_seconds() > 10 * 60:
            2/0
    
#     if p1_deck:
#         print(f'The winner of game {game} is player 1!')
#     else:
#         print(f'The winner of game {game} is player 2!')
        
    if game > 1:
        return P1_WINS if p1_deck else P2_WINS
    return p1_deck if p1_deck else p2_deck

    

wd = play_game(deepcopy(p1_deck), deepcopy(p2_deck), 1, set())
wd, sum([
    (n+1) * i 
    for n, i in enumerate(reversed(wd))
])
```
