```python
import copy


def adjacent(state, row, col):
    r = []
    
    last_col_idx = len(state[0]) - 1
    last_row_idx = len(state) - 1
    
    if row > 0:
        if col > 0:
            r.append(state[row-1][col-1])
        r.append(state[row-1][col])
        if col < last_col_idx:
            r.append(state[row-1][col+1])
            
    if col > 0:
        r.append(state[row][col-1])
    if col < last_col_idx:
        r.append(state[row][col+1])
        
    if row < last_row_idx:
        if col > 0:
            r.append(state[row+1][col-1])
        r.append(state[row+1][col])
        if col < last_col_idx:
            r.append(state[row+1][col+1])

    return r
    
    

def step(state):
    init = copy.deepcopy(state)
    new = copy.deepcopy(state)
    
    ops = 0
    
    for row, line in enumerate(init):
        for col, c in enumerate(line):
            adj = adjacent(init, row, col)
            
            if c == 'L' and not (set(adj) & {'#'}):
                new[row][col] = '#'
                ops += 1
            if c == '#' and adj.count('#') > 3:
                new[row][col] = 'L'
                ops += 1
                
    return new, ops
              

def viz(state):
    return '\n'.join([''.join(i) for i in state])


# print(viz(inp))
# print()
# print(viz(step(step(inp))))

steps = 0
state = copy.deepcopy(inp)
while True:
    steps += 1
    new_state, ops = step(state)
    state = new_state
    if ops == 0:
        break
    

# print(viz(state))
print(steps)
print(viz(state).count('#'))
```

```python
def adjacent2(state, row, col):
    r = []
    
    last_col_idx = len(state[0]) - 1
    last_row_idx = len(state) - 1
    seeking = {'#', 'L'}
    
    def in_bounds(row, col):
        if row < 0 or row > last_row_idx:
            return False
        if col < 0 or col > last_col_idx:
            return False
        
        return True

    
    def nearest(row, col, diffrow, diffcol):
        other_row, other_col = row, col
        for _ in range(1000):
            other_row += diffrow
            other_col += diffcol
            
            if not in_bounds(other_row, other_col):
                return None
            
            c = state[other_row][other_col]
            if c in seeking:
                return c
            
    
    for c in [
        nearest(row, col, -1, -1),
        nearest(row, col, -1, 0),
        nearest(row, col, -1, 1),
        nearest(row, col, 0, -1),
        nearest(row, col, 0, 1),
        nearest(row, col, 1, -1),
        nearest(row, col, 1, 0),
        nearest(row, col, 1, 1),
    ]:
        if c:
            r.append(c)

    return r
    
    

def step2(state):
    init = copy.deepcopy(state)
    new = copy.deepcopy(state)
    
    ops = 0
    
    for row, line in enumerate(init):
        for col, c in enumerate(line):
            adj = adjacent2(init, row, col)
            
            if c == 'L' and not (set(adj) & {'#'}):
                new[row][col] = '#'
                ops += 1
            if c == '#' and adj.count('#') > 4:
                new[row][col] = 'L'
                ops += 1
                
    return new, ops
              

# print(viz(inp))
# print()
# print(viz(step2(inp)[0]))
# print()
# print(viz(step2(step2(inp)[0])[0]))

steps = 0
state = copy.deepcopy(inp)
while True:
    steps += 1
    new_state, ops = step2(state)
    state = new_state
    if ops == 0:
        break
    

# print(viz(state))
print(steps)
print(viz(state).count('#'))
```
