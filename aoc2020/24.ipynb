```python
def get_directions(lines):
    res = []
    for line in lines:
        directions = []
        while line:
            for d in [
                'se',
                'sw',
                'ne',
                'nw',
                'e',
                'w',
            ]:
                if line.startswith(d):
                    directions.append(d)
                    line = line[len(d):]
                    break
        res.append(directions)
    return res



flipped = {}

for line in get_directions(inp):
    x, y, z = 0, 0, 0

    for direction in line:
        update = {
            'se': (0, -1, 1),
            'sw': (-1, 0, 1),
            'ne': (1, 0, -1),
            'nw': (0, 1, -1),
            'e':  (1, -1, 0),
            'w':  (-1, 1, 0),

        }[direction]
        x += update[0]
        y += update[1]
        z += update[2]
    
    key = x, y, z
    if key in flipped:
#         print('flipping again', key)
        flipped.pop(key)
    else:    
        flipped[key] = 1
    
len(flipped)
```

```python
def get_adjacent(x, y, z):
    return [
        (x-1, y+1, z),  # w
        (x-1, y, z+1),  # sw
        (x, y-1, z+1),  # se
        (x+1, y-1, z),  # e
        (x+1, y, z-1),  # ne
        (x, y+1, z-1),  # nw
    ]
    

def day(flipped):
    new = deepcopy(flipped)
#     print('day starts', sorted(flipped.keys()))

    # Any black tile with zero or more than 2 black tiles 
    # immediately adjacent to it is flipped to white.    
    for black in flipped:
        adj_black = [
            adj for adj in get_adjacent(*black) 
            if adj in flipped
        ]
        if not adj_black or len(adj_black) > 2:
#             print('flipping to white', black, adj_black)
            new.pop(black)
            
    # Any white tile with exactly 2 black tiles 
    # immediately adjacent to it is flipped to black.
    for black in flipped:
        adj_white_tiles = [adj for adj in get_adjacent(*black) if adj not in flipped]
        
        for white in adj_white_tiles:
            adj_black_tiles = [adj for adj in get_adjacent(*white) if adj in flipped]
            if len(adj_black_tiles) == 2:
#                 print('flipping to black', white)
                new[white] = 1
        
    return new
    
day_iter = flipped
for t in range(1, 101):
    day_iter = day(day_iter)
    print(f'Day {t}: {len(day_iter)}')
```
